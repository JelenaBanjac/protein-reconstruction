```python
import os
import h5py
from sklearn.neighbors import NearestNeighbors
import numpy as np   
import ipyvolume as ipv
import time

import sys
sys.path.append("..") 
from cryoem.plots import plot_projection, plot_projections
from cryoem.projections import RotationMatrix
```

## Euler Angles

```python
angle_ranges = "1.0,0.0,0.0"
angle_shift = "0.0,0.0,0.0"
projections_filename = "../data/bgal_ProjectionsAngles_ProjNber20_AngCoverage1.0,0.0,0.0_AngShift0.0,0.0,0.0.h5"#f"../data/5j0n_ProjectionsAngles_ProjNber5000_AngCoverage{angle_ranges}_AngShift{angle_shift}.h5"

# load structures
data = h5py.File(projections_filename, 'r')

# plot
all_vectors = RotationMatrix(data["Angles"])
ipv.figure(width=500, height=500)
ipv.pylab.xlim(-1, 1)
ipv.pylab.ylim(-1, 1)
ipv.pylab.zlim(-1, 1)
ipv.scatter(all_vectors[:,0], all_vectors[:,2], all_vectors[:,1], marker="sphere", color="blue", size=1)
ipv.pylab.save(f"data/angle_variety/{angle_ranges}.html")
ipv.show()
```
